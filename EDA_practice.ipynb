```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("vaccination-data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['WHO_REGION']
```

```python
common_region = df["WHO_REGION"].mode()[0]
df['WHO_REGION'].fillna(common_region,inplace = True)
```

```python
df['DATE_UPDATED']
```

```python
df.dropna(subset=['DATE_UPDATED'],inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum() / len(df)
```

```python
df.drop(['VACCINES_USED','NUMBER_VACCINES_TYPES_USED'], axis=1, inplace=True)
```

```python
df.isna().sum() / len(df)
```

```python
m1=df['TOTAL_VACCINATIONS_PER100'].mean()
df['TOTAL_VACCINATIONS_PER100'].fillna(m1,inplace=True)
```

```python
m2=df['PERSONS_VACCINATED_1PLUS_DOSE_PER100'].mean()
df['PERSONS_VACCINATED_1PLUS_DOSE_PER100'].fillna(m2,inplace=True)
```

```python
m3=df['PERSONS_LAST_DOSE_PER100'].mean()
df['PERSONS_LAST_DOSE_PER100'].fillna(m3,inplace=True)
```

```python
m4=df['PERSONS_BOOSTER_ADD_DOSE_PER100'].mean()
df['PERSONS_BOOSTER_ADD_DOSE_PER100'].fillna(m4,inplace=True)
```

```python
df.isna().sum() / len(df)
```

```python
df.iloc[:,-3:]   # to view the last three
```

```python
df['PERSONS_BOOSTER_ADD_DOSE'].median()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['PERSONS_BOOSTER_ADD_DOSE'])
plt.title('Plot of PERSONS_BOOSTER_ADD_DOSE')
plt.xlabel('Index')
plt.ylabel('PERSONS_BOOSTER_ADD_DOSE')
plt.grid(True)
plt.show()
```

```python
median_value = df['PERSONS_BOOSTER_ADD_DOSE'].median()
df['PERSONS_BOOSTER_ADD_DOSE'].fillna(median_value, inplace=True)
```

```python
# after cleaning
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['PERSONS_BOOSTER_ADD_DOSE'])
plt.title('Plot of PERSONS_BOOSTER_ADD_DOSE')
plt.xlabel('Index')
plt.ylabel('PERSONS_BOOSTER_ADD_DOSE')
plt.grid(True)
plt.show()
```
